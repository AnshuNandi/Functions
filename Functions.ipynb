1. What is the difference between a function and a method in Python?

Answer:

- A function in Python is a block of reusable code defined independently using the def keyword or lambda. It can be called directly without being tied to an object.
- A method is also a function but it is defined inside a class and is associated with an object. It often operates on the objectâ€™s data.

```python
#Example
def add(a, b):
    return a + b   # Function

class Test:
    def show(self):
        return "Hello"   # Method
```

2. Explain the concept of function arguments and parameters in Python.

Answer:

- Parameter: Variable in the function definition.

- Argument: Actual value passed during function call.

```python
#example
def greet(name):  # 'name' is parameter
    return "Hello " + name

print(greet("Anshu"))  # "Anshu" is argument
```

5. What are iterators in Python and how do they differ from iterables?

Answer:

- Iterable: Any object we can loop over (list, tuple, str).

- Iterator: An object that produces values one at a time using __next__().

```python
#Example
nums = [10, 20, 30]   # iterable
it = iter(nums)       # creates an iterator from the list
print(next(it))
print(next(it))
print(next(it))
```

6. Explain the concept of generators in Python and how they are defined.

Answer: A generator is a special function that uses yield to produce values one at a time. It does not store all values in memory.

```python
#Example
def g():
    for i in range(3): yield i
for x in g(): print(x)
```

7. What are the advantages of using generators over regular functions?

Answer:

- More memory efficient (no need to store large data structures).

- Support lazy evaluation (values generated only when needed).

- Can represent infinite sequences.

8. What is a lambda function in Python and when is it typically used?

Answer:
A lambda function is a small, anonymous, one-line function.
It is often used where short-lived functions are needed (e.g., sorting, mapping, filtering).

```python
#Example
f=lambda a,b:a+b
print(f(3,4))
```

9. Explain the purpose and usage of the map() function in Python.

Answer:
The map() function applies a given function to each element of an iterable and returns a new iterable (map object).

```python
#Example:
lst=[1,2,3]
print(list(map(lambda x:x*2,lst)))
```

10. What is the difference between map(), reduce(), and filter() functions in Python?

Answer:

- map(): Applies a function to all items of an iterable.

- filter(): Keeps items that satisfy a condition.

- reduce(): Repeatedly applies a function to pairs of items, reducing iterable to a single value.

```python
#Example:
from functools import reduce
lst=[1,2,3,4]
print(list(map(lambda x:x*2,lst)))    # [2,4,6,8]
print(list(filter(lambda x:x%2==0,lst))) # [2,4]
print(reduce(lambda a,b:a+b,lst))  # 10
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def s(lst):
    return sum([x for x in lst if x%2==0])

print(s([1,2,3,4,5,6]))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def p(n):
    if n<2:return 0
    for i in range(2,int(n**0.5)+1):
        if n%i==0:return 0
    return 1

print([i for i in range(1,21) if p(i)])
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class F:
    def __init__(s,n):s.n=n;s.a=0;s.b=1;s.i=0
    def __iter__(s):return s
    def __next__(s):
        if s.i>=s.n:raise StopIteration
        if s.i==0:r=0
        elif s.i==1:r=1
        else:r=s.a+s.b;s.a,s.b=s.b,r
        s.i+=1
        return r

print(list(F(7)))
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def g(n):
    for i in range(n+1):yield 2**i

print(list(g(5)))
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]

def g(f):
    with open(f) as fp:
        for l in fp:
            yield l.strip()

for line in g(filename):
    print(line)
```
